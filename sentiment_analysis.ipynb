```python
import pandas as pd
import numpy as np
```

```python
# convert to dataframe
df = pd.DataFrame(np.array(reviews), columns = ['text'])
```

```python
df
```

# Text cleaning

```python
# removing punctuations
df['text_clean'] = df['text'].str.replace('[^\w\s]', '')
```

```python
# \n
df['text_cleaned'] = df['text_clean'].str.replace('\n', ' ')
```

```python
df
```

```python
# lowercase all text
df['text_lowercase'] = df['text_cleaned'].apply(lambda x: ' '.join(x.lower() for x in x.split()))
```

```python
df
```

```python
# remove stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_corpus = stopwords.words('english')
```

```python
df['text_removed_stopwords'] = df['text_lowercase'].apply(lambda x: ' '.join(x for x in x.split() if x not in stopwords_corpus))
```

```python
freq_words = pd.Series(' '.join(df['text_removed_stopwords']).split()).value_counts()[:10]
```

```python
freq_words
```

# Lemmatize reviews

```python
from textblob import Word
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
df['clean_review'] = df['text_removed_stopwords'].apply(lambda x: ' '.join([Word(word).lemmatize() for word in x.split()]))
```

```python
# will keep only needed columns
df = df.drop(['text_clean', 'text_cleaned', 'text_lowercase', 'text_removed_stopwords'], axis = 1)
```

# Sentiment analysis TextBlob

```python
# calculate review score
# polarity between 1;-1
# subjectivity between 0;1
from textblob import TextBlob
```

```python
df['polarity'] = df['clean_review'].apply(lambda x: TextBlob(x).sentiment[0])
```

```python
df['subjectivity'] = df['clean_review'].apply(lambda x: TextBlob(x).sentiment[1])
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
# initiate models
# score between 1:5
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')

model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
# test
tokens = tokenizer.encode('It was good but couldve been better. Great', return_tensors='pt')
```

```python
result = model(tokens)
```

```python
result.logits
```

```python
int(torch.argmax(result.logits))+1
```

```python
df
```

```python
def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
df['sentiment'] = df['text'].apply(lambda x: sentiment_score(x[:512]))
```

```python
df['sentiment'] = df['clean_review'].apply(lambda x: sentiment_score(x[:512]))
```
